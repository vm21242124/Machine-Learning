```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
dataset
```

```python
x=dataset.iloc[:,1:-1].values
x
```

```python
y=dataset.iloc[:,-1].values
y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)    #training the model on the data
```

```python
from sklearn.preprocessing import PolynomialFeatures
pl_reg= PolynomialFeatures(degree=4)
x_poly=pl_reg.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title("Truth or mulf (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(pl_reg.fit_transform(x)),color='blue')
plt.title("Truth or mulf (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(pl_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

** Predicting a new result with Linear Regression**

```python
lin_reg.predict([[6.5]])
```

** Predicting a new result with Polynomial Regression**

```python
lin_reg_2.predict(pl_reg.fit_transform([[6.5]]))
```
